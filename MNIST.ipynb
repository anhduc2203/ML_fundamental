```python
X = mnist['data']
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
x_train, x_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
ann = MLPClassifier(hidden_layer_sizes=(100,), solver='sgd', activation='relu', alpha=0.0001, verbose=20, batch_size=200, learning_rate_init=0.0001, max_iter=100, tol=0.0001, random_state=42)
```

```python
import numpy as np
```

```python
shuffle_index = np.random.permutation(60000)
```

```python
x_train, y_train = x_train[shuffle_index], y_train[shuffle_index]
```

```python
ann.fit(x_train, y_train)
```

```python
print("Do chinh xac traing: {}".format(ann.score(x_train, y_train)))
```

```python
print("Do chinh xac testing: {}".format(ann.score(x_test, y_test)))
```

```python
vmin, vmax = ann.coefs_[0].min(), ann.coefs_[0].max()
```
